## Introdução
O projeto tem como finalidade analisar como e qual é o tipo de acesso à internet da população brasileira, de acordo com sua renda e localização. Para isso, a base dos dados será a PNAD de 2014, na qual analisaremos as seguintes variáveis: tem acesso à Internet no próprio domicilio, o acesso à Internet é feito através de sinal de rede celular 3G ou 4G, o acesso à Internet é feito através de conexão discada por linha telefônica, o acesso à Internet é feito por banda larga, faixa do rendimento mensal domiciliar per capita e código de situação censitária.

## Análise e Resultados

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
print('Esperamos trabalhar no diretório')
print(os.getcwd())
```

```python
dados = pd.read_excel('DOM2014.xlsx')
```

```python
df = pd.DataFrame({'Internet': dados.V02322,
                   'd3Gou4G': dados.V02424,
                   'Bandalarga': dados.V02426,
                   'Discada': dados.V02425,
                   'Região': dados.V4105,
                   'Renda': dados.V4622})
df
```

```python
dfn = df.dropna()
dfn
```

```python
df.d3Gou4G = df.d3Gou4G.astype('category')
df.d3Gou4G.cat.categories = (['Sim', 'Não'])
df.Bandalarga = df.Bandalarga.astype('category')
df.Bandalarga.cat.categories = (['Sim', 'Não'])
df.Discada = df.Discada.astype('category')
df.Discada.cat.categories = (['Sim', 'Não'])
df.Internet = df.Internet.astype('category')
df.Internet.cat.categories = (['Sim', 'Não'])
df.Região = df.Região.astype('category')
df.Região.cat.categories = (['Urbana','Urbana1','Urbana2','Rural','Rural1','Rural2','Rural3','Rural4'])
df.Renda = df.Renda.astype('category')
df.Renda.cat.categories = (['Sem rendimento', 'Até ¼ salário mínimo','Mais de ¼ até ½ salário mínimo','Mais de ½ até 1 salário mínimo','Mais de 1 até  2 salários mínimos','Mais de  2 até  3 salários mínimos','Mais de 3 até 5 salários mínimos','Mais de 5 salários mínimos','vazio'])

df.dtypes
```

```python
#Uso da Internet Banda Larga
df1 = df['Bandalarga'].value_counts()
df1.plot(kind='pie', colors=('yellow', 'salmon'), autopct="%0.2f",legend=True)
plt.title('Uso da Internet Banda Larga')
```

```python
#Uso da Internet Discada
df2 = df['Discada'].value_counts()
df2.plot(kind='pie', colors=('yellow', 'salmon'), autopct="%0.2f",legend=True)
plt.title('Uso da Internet Discada')
```

```python
#Uso da Internet 3G ou 4G
df3 = df['d3Gou4G'].value_counts()
df3.plot(kind='pie', colors=('yellow', 'salmon'), autopct="%0.2f",legend=True)
plt.title('Uso da Internet 3G ou 4G')
```

```python
#Internet x Renda
ct1 =pd.crosstab(df['Renda'],df['Internet'])
ct1 = (ct1/ct1.sum().sum()*100).round(decimals=0)
ct1.plot(kind='bar', color=('yellow','salmon'))
plt.ylabel('Frequência Relativa (em %)')
plt.title('Acesso à internet por faixa de renda')
plt.xlim(-0.5,7.5)
```

```python
#Internet Discada x Renda
ct2 = pd.crosstab(df['Renda'], df['Discada'])
ct2.plot(kind='bar', color=('yellow', 'salmon'))
plt.ylabel('Frequência Absoluta')
plt.xlim(-0.5,7.5)
plt.title('Internet Discada de acordo com a renda')
```

```python
#Internet Banda Larga x Renda
ct3 = pd.crosstab(df['Renda'], df['Bandalarga'])
ct3.plot(kind='bar', color=('yellow', 'salmon'))
plt.ylabel('Frequência Absoluta')
plt.xlim(-0.5,7.5)
plt.title('Internet Banda Larga de acordo com a renda')
```

```python
#Internet 3G ou 4G x Renda
ct4 = pd.crosstab(df['Renda'], df['d3Gou4G'])
ct4.plot(kind='bar', color=('yellow', 'salmon'))
plt.ylabel('Frequência Absoluta')
plt.xlim(-0.5,7.5)
plt.title('Internet 3G ou 4G de acordo com a renda')
```

```python
#Internet Região x Internet
ct5 =pd.crosstab(df['Região'],df['Internet'])
ct5 = (ct5/ct5.sum().sum()*100).round(decimals=0)
ct5.plot(kind='bar', color=('yellow','salmon'))
plt.ylabel('Frequência Relativa (em %)')
plt.title('Acesso à internet por região')
```

```python
#Internet Discada x Região
ct6 = pd.crosstab(df['Região'], df['Discada'])
ct6.plot(kind='bar', color=('yellow', 'salmon'))
plt.ylabel('Frequência Absoluta')
plt.xlim(-0.5,7.5)
plt.title('Uso da internet Discada por região')
```

```python
#Internet Banda Larga x Região
ct7 = pd.crosstab(df['Região'], df['Bandalarga'])
ct7.plot(kind='bar', color=('yellow', 'salmon'))
plt.ylabel('Frequência Absoluta')
plt.xlim(-0.5,7.5)
plt.title('Uso da internet Banda Larga por região')
```

```python
#Internet 3g ou 4G x Região
ct8 = pd.crosstab(df['Região'], df['d3Gou4G'])
ct8.plot(kind='bar', color=('yellow', 'salmon'))
plt.ylabel('Frequência Absoluta')
plt.xlim(-0.5,7.5)
plt.title('Uso da internet  3G ou 4G por região')
```

## Conclusão 

A partir das análises que foram feitas é possível concluir que a área urbana tem muito mais acesso à internet do que a área rural. Isso mostra o quanto a desigualdade de informação pode ser prejudicial para o desenvolvimento da sociedade, já que a internet permite não só a conexão entre pessoas, mas também o acesso a informações básicas para a população. Por exemplo, pela internet é possível acessar o site da prefeitura da cidade, que possibilita o acompanhamento das leis e projetos em andamento e a participação na política da sua cidade, exercendo a cidadania. Porém, pouquíssimos moradores das áreas rurais fazem uso dessa tecnologia, e quando fazem utilizam a internet 3G ou 4G, que não proporciona a melhor qualidade de conexão. Além disso, pode-se analisar também que a internet discada, que há 10 anos era a única opção de conexão, deixou de ser utilizada pelos brasileiros, já que, nem mesmo nas áreas rurais ela tem grande expressividade. Acerca da tecnologia 3G e 4G, podemos avaliar que elas estão em um crescente, visto que, a maioria das pessoas já utilizam esse tipo de internet, porém uma parcela considerável ainda não faz uso desse produto tecnológico, utilizando a internet banda larga como principal fonte de conexão.
